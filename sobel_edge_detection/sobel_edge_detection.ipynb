```python
def convert_rgb_to_gray_level(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]
    im_2=np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            im_2[i,j]=get_distance(im_1[i,j,:])
    return im_2
def get_distance(v,w=[1/3,1/3,1/3]):
    a,b,c=v[0],v[1],v[2]
    w1,w2,w3=w[0],w[1],w[2]
    d= ((a**2)*w1 + (b**2)*w2 + (c**2)*w3)**.5
    # d=((a*w1)**2+(b*w2)**2+(c*w3)**2)**.5
    return d
```
